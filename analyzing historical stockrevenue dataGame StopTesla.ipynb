```python
import yfinance as yf
import pandas as pd

# Import Stock Data Tesla
tesla_data = yf.download('TSLA', start='2010-01-01', end='2023-12-31')
tesla_data.head()
```

```python
# Install requests dan BeautifulSoup jika belum terpasang
!pip install requests beautifulsoup4

# Mengimpor pustaka yang dibutuhkan
import requests
from bs4 import BeautifulSoup

# Mengambil halaman web (ganti URL_WEBSITE_DENGAN_DATA_PENDAPATAN_TESLA dengan URL sebenarnya)
url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Mengambil data pendapatan dari tabel
tables = soup.find_all('table')
tesla_revenue = pd.read_html(str(tables))[1]  # Mengambil tabel kedua
tesla_revenue = tesla_revenue.iloc[::-1]  # Membalik urutan data
tesla_revenue.columns = ['Date', 'Revenue']
tesla_revenue = tesla_revenue[tesla_revenue['Revenue'] != '']
tesla_revenue['Revenue'] = tesla_revenue['Revenue'].str.replace('$', '').str.replace(',', '').astype(float)
tesla_revenue.head()


```

```python
# Mengambil data saham GameStop
gamestop_data = yf.download('GME', start='2010-01-01', end='2023-12-31')
gamestop_data.head()
```

```python
# Mengambil halaman web (ganti URL_WEBSITE_DENGAN_DATA_PENDAPATAN_GAMESTOP dengan URL sebenarnya)
url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Mengambil data pendapatan dari tabel
tables = soup.find_all('table')
gamestop_revenue = pd.read_html(str(tables))[1]  # Mengambil tabel kedua
gamestop_revenue = gamestop_revenue.iloc[::-1]  # Membalik urutan data
gamestop_revenue.columns = ['Date', 'Revenue']
gamestop_revenue = gamestop_revenue[gamestop_revenue['Revenue'] != '']
gamestop_revenue['Revenue'] = gamestop_revenue['Revenue'].str.replace('$', '').str.replace(',', '').astype(float)
gamestop_revenue.head()
```

```python
import matplotlib.pyplot as plt

# Grafik harga saham Tesla
plt.figure(figsize=(10, 5))
plt.plot(tesla_data['Close'], label='Tesla Stock Price')
plt.title('Tesla Stock Price Over Time')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.show()

# Grafik pendapatan Tesla
plt.figure(figsize=(10, 5))
plt.plot(tesla_revenue['Date'], tesla_revenue['Revenue'], label='Tesla Revenue')
plt.title('Tesla Revenue Over Time')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.legend()
plt.show()
```

```python
# Grafik harga saham GameStop
plt.figure(figsize=(10, 5))
plt.plot(gamestop_data['Close'], label='GameStop Stock Price')
plt.title('GameStop Stock Price Over Time')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.show()

# Grafik pendapatan GameStop
plt.figure(figsize=(10, 5))
plt.plot(gamestop_revenue['Date'], gamestop_revenue['Revenue'], label='GameStop Revenue')
plt.title('GameStop Revenue Over Time')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.legend()
plt.show()
```
